```python
import seaborn as sns
from ipywidgets import interact
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
@interact(hue=['smoker', 'sex', 'time', 'day'])
def plot(hue):
    _ = sns.pairplot(tips, hue=hue)
```
